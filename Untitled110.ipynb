```python
# ===============================
# CNN FEATURE EXTRACTION NOTEBOOK
# ===============================
# COMPLETE NOTEBOOK CODE IN ONE CELL BLOCK
# ===============================

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np
import os
from datetime import datetime

# -------------------------------
# SETUP & CONFIG
# -------------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

torch.manual_seed(42)

# Create directory for images
if not os.path.exists("images"):
    os.makedirs("images")


# -------------------------------
# DATA LOADING (MNIST)
# -------------------------------
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])

train_set = datasets.MNIST(root="./data", train=True, download=True, transform=transform)
test_set  = datasets.MNIST(root="./data", train=False, download=True, transform=transform)

train_loader = DataLoader(train_set, batch_size=64, shuffle=True)
test_loader  = DataLoader(test_set, batch_size=1000, shuffle=False)


# -------------------------------
# MODEL DEFINITON (CNN)
# -------------------------------
class SimpleCNN(nn.Module):
    def __init__(self):
        super().__init__()
        # Conv layers
        self.conv1 = nn.Conv2d(1, 8, kernel_size=3, padding=1)   # 1 → 8 filters
        self.conv2 = nn.Conv2d(8, 16, kernel_size=3, padding=1)  # 8 → 16 filters

        # Fully connected
        self.fc1 = nn.Linear(16 * 7 * 7, 128)
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        x = F.relu(self.conv1(x))  # Feature maps 1
        x = F.max_pool2d(x, 2)
        x = F.relu(self.conv2(x))  # Feature maps 2
        x = F.max_pool2d(x, 2)
        x = torch.flatten(x, 1)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x


model = SimpleCNN().to(device)
optimizer = optim.Adam(model.parameters(), lr=0.001)
criterion = nn.CrossEntropyLoss()


# -------------------------------
# TRAINING LOOP
# -------------------------------
epochs = 3
train_losses = []
train_acc = []

print("Training started...")

for epoch in range(epochs):
    model.train()
    total = 0
    correct = 0
    running_loss = 0

    for data, target in train_loader:
        data, target = data.to(device), target.to(device)

        optimizer.zero_grad()
        output = model(data)
        loss = criterion(output, target)

        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, pred = torch.max(output.data, 1)
        total += target.size(0)
        correct += (pred == target).sum().item()

    epoch_loss = running_loss / len(train_loader)
    epoch_acc = 100 * correct / total

    train_losses.append(epoch_loss)
    train_acc.append(epoch_acc)
    print(f"Epoch {epoch+1}/{epochs}   Loss: {epoch_loss:.4f}   Accuracy: {epoch_acc:.2f}%")


# -------------------------------
# PLOT TRAINING CURVES
# -------------------------------
plt.figure(figsize=(10,4))
plt.plot(train_losses)
plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.savefig("images/training_loss.png", dpi=200, bbox_inches="tight")
plt.show()

plt.figure(figsize=(10,4))
plt.plot(train_acc)
plt.title("Training Accuracy (%)")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.savefig("images/training_accuracy.png", dpi=200, bbox_inches="tight")
plt.show()


# -------------------------------
# VISUALISE FILTERS OF FIRST LAYER
# -------------------------------
filters = model.conv1.weight.data.clone().cpu()

fig, ax = plt.subplots(2, 4, figsize=(8,4))
ax = ax.flatten()

for i in range(8):
    ax[i].imshow(filters[i,0,:,:], cmap="gray")
    ax[i].axis("off")

plt.suptitle("Conv1 Filters (8 Filters)")
plt.savefig("images/conv1_filters.png", dpi=200, bbox_inches="tight")
plt.show()


# -------------------------------
# FEATURE MAP EXTRACTION USING HOOKS
# -------------------------------
feature_maps = {}

def hook_fn(name):
    def hook(module, input, output):
        feature_maps[name] = output.detach().cpu()
    return hook

# Attach hooks
model.conv1.register_forward_hook(hook_fn("conv1"))
model.conv2.register_forward_hook(hook_fn("conv2"))

# Pass a sample through the network
sample_img, _ = test_set[0]
sample = sample_img.unsqueeze(0).to(device)
_ = model(sample)


# -------------------------------
# VISUALISE FEATURE MAPS FOR CONV1
# -------------------------------
fm1 = feature_maps["conv1"][0]

fig, ax = plt.subplots(2, 4, figsize=(8,4))
ax = ax.flatten()
for i in range(8):
    ax[i].imshow(fm1[i], cmap="gray")
    ax[i].axis("off")

plt.suptitle("Feature Maps: Conv1 (After ReLU)")
plt.savefig("images/featuremaps_conv1.png", dpi=200, bbox_inches="tight")
plt.show()


# -------------------------------
# VISUALISE FEATURE MAPS FOR CONV2
# -------------------------------
fm2 = feature_maps["conv2"][0]

fig, ax = plt.subplots(4, 4, figsize=(8,8))
ax = ax.flatten()
for i in range(16):
    ax[i].imshow(fm2[i], cmap="gray")
    ax[i].axis("off")

plt.suptitle("Feature Maps: Conv2 (After ReLU)")
plt.savefig("images/featuremaps_conv2.png", dpi=200, bbox_inches="tight")
plt.show()


# -------------------------------
# TEST ACCURACY
# -------------------------------
model.eval()
correct = 0
total = 0

with torch.no_grad():
    for data, target in test_loader:
        data, target = data.to(device), target.to(device)
        output = model(data)
        _, pred = torch.max(output.data, 1)
        total += target.size(0)
        correct += (pred == target).sum().item()

print(f"Final Test Accuracy: {100 * correct/total:.2f}%")

print("\nAll images saved in /images folder.")
```
